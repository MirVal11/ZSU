## ðŸ”Ž Can you vizualize what was the average price of the house by quarters?

* We need to create a new column **YearQuarterSold** with merged Year and Quarter information in this pattern: '2010-1','2010-2' and so on first
    * We can map the values using the *apply* function or use string concatenation directly
* ðŸ’¡ If you need to change data type of the columns, you can use *astype*

```python
df_full["QuarterSold"] = (df_full.MoSold + 2) // 3
df_full["YearQuarterSold"] = df_full.apply(lambda row: f"{row.YrSold}-{row.QuarterSold}", axis=1)
df_full["YearQuarterSold"]
```

### Compute the average price for the *YearQuarterSold* attribute

```python
df_agg = df_full.SalePrice.groupby(df_full.YearQuarterSold).mean().apply(np.round).astype(np.int64).reset_index(name="MeanSalePrice")
```

```python
plt.xticks(rotation=45, horizontalalignment="right")
sns.lineplot(df_agg, x="YearQuarterSold", y="MeanSalePrice")
```

```python
Q1_SalePrice = df_full['SalePrice'].quantile(0.25)
Q3_SalePrice = df_full['SalePrice'].quantile(0.75)
IQR_SalePrice = Q3_SalePrice - Q1_SalePrice

Q1_GrLivArea = df_full['GrLivArea'].quantile(0.25)
Q3_GrLivArea = df_full['GrLivArea'].quantile(0.75)
IQR_GrLivArea = Q3_GrLivArea - Q1_GrLivArea

lower_bound_SalePrice = Q1_SalePrice - 1.5 * IQR_SalePrice
upper_bound_SalePrice = Q3_SalePrice + 1.5 * IQR_SalePrice

lower_bound_GrLivArea = Q1_GrLivArea - 1.5 * IQR_GrLivArea
upper_bound_GrLivArea = Q3_GrLivArea + 1.5 * IQR_GrLivArea

df_full['Outlier'] = ((df_full['SalePrice'] < lower_bound_SalePrice) |
                      (df_full['SalePrice'] > upper_bound_SalePrice) |
                      (df_full['GrLivArea'] < lower_bound_GrLivArea) |
                      (df_full['GrLivArea'] > upper_bound_GrLivArea))

plt.figure(figsize=(12, 8))
sns.scatterplot(x=df_full['GrLivArea'], y=df_full['SalePrice'], hue=df_full['Outlier'])
plt.title('Outliers in SalePrice and GrLivArea')
plt.show()
```
